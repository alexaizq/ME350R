```python
# import required libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.animation import FuncAnimation

data = pd.read_csv('motiongen-output.csv')
data.describe()
```

```python
# plot x vs y
fig, ax = plt.subplots()
x = data['x [in]']
y = data['y [in]']

line, = ax.plot(x, y)

ax.grid(True)
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.title('Position vs Time')
plt.show()
```

```python
# plot the displacement over time from data
fig, ax = plt.subplots()
x = data['Time [s]']
y = data['Displacement [in]']

line, = ax.plot(x, y)

ax.grid(True)
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.title('Displacement vs Time')
plt.show()
```

```python
# plot velocity vs time
fig, ax = plt.subplots()
x = data['Time [s]']
y = data['Linear Vel. [in/s]']

line, = ax.plot(x, y)

ax.grid(True)
ax.set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.title('Velocity vs Time')
plt.show()
```

```python
# plot acceleration vs time
fig, ax = plt.subplots()
x = data['Time [s]']
y = data['Linear Accel. [in/s^2]']

line, = ax.plot(x, y)

ax.grid(True)
# ax.set_aspect('equal', adjustable='box')
# plt.tight_layout()
plt.title('Acceleration vs Time')
plt.show()
```
